### season

```python
import re
import pandas as pd

# Define common season terms (adjustable)
# season_keywords = r'весен|летн|осен|зимн|тепл|деми'
season_keywords = r'весен|летн|осен|зимн' # explicit season only

def first_two_sentences(text):
    """Extract the first two sentences from the given text."""
    # Ensure text is a string and split by punctuation followed by whitespace
    sentences = re.split(r'(?<=[.!?])\s+', str(text))
    # Return the first two sentences joined together
    return " ".join(sentences[:2])

# Create a boolean mask by applying regex search on the first two sentences
mask = wb_oz_df['WBDescription'].apply(
    lambda x: bool(re.search(season_keywords, first_two_sentences(x), flags=re.IGNORECASE))
)

season_mentions = wb_oz_df[mask]
print(f'total season_mentions: {len(season_mentions)}')

pd.set_option('display.max_colwidth', None)
display(
    season_mentions['WBDescription'].sample(1)
)
pd.reset_option('display')
```

```python
import re
import pandas as pd
import inspect

# Fix for python 3.13
import inspect

def getargspec_wrapper(func):
    full = inspect.getfullargspec(func)
    # Return only the args, varargs, varkw, and defaults
    return (full.args, full.varargs, full.varkw, full.defaults)

# if not hasattr(inspect, "getargspec"):
inspect.getargspec = getargspec_wrapper

import pymorphy2

# Initialize the morphological analyzer
morph = pymorphy2.MorphAnalyzer()

# Define the season keywords in their base (lemma) forms
season_keywords = ['весна', 'лето', 'осень', 'зима']

def text_mentions_season(text, morph, season_keywords):
    """
    Check if the input text contains any season keywords,
    accounting for different word forms using pymorphy2.
    """
    # Ensure text is a string and convert to lowercase
    text = str(text).lower()
    # Extract word tokens using regex (ignoring punctuation)
    tokens = re.findall(r'\w+', text)
    for token in tokens:
        # Get the lemma (normal form) of the token
        lemma = morph.parse(token)[0].normal_form
        if lemma in season_keywords:
            return True
    return False

# Apply the function to filter rows in the 'WBDescription' column
mask = wb_oz_df['WBDescription'].apply(lambda x: text_mentions_season(x, morph, season_keywords))
season_mentions = wb_oz_df[mask]
print(f"total season_mentions: {len(season_mentions)}")
```
