```python
# Clean product name

import re

def clean_product_name_full(row):
    name = row['OZProductName (ProductName)']
    brand = str(row['OZBrandName (OZBrandName)'])

    # Remove brand name (raw and in parentheses)
    if brand:
        brand_pattern = re.escape(brand)
        name = re.sub(brand_pattern, '', name, flags=re.IGNORECASE)
        name = re.sub(r'\(\s*' + brand_pattern + r'\s*\)', '', name, flags=re.IGNORECASE)

    # Remove any brand-like patterns in parentheses (e.g. (РАССВЕТ))
    name = re.sub(r'\(\s*[^)]*\s*\)', '', name)

    # Remove size/quantity patterns
    size_patterns = [
        r'размер\s*\d+[,\d\sшт]*',
        r'\b\d+\s*(шт)?\b',
    ]
    for pat in size_patterns:
        name = re.sub(pat, '', name, flags=re.IGNORECASE)

    # Remove discount/marketing messages
    discount_keywords = [
        r'\bскидка\b',
        r'на последний размер',
        r'\bакция\b',
        r'\bпромо\b'
    ]
    for pat in discount_keywords:
        name = re.sub(pat, '', name, flags=re.IGNORECASE)

    # Clean up spacing and punctuation
    name = re.sub(r'\s+', ' ', name)         # collapse multiple spaces
    name = re.sub(r'[,\s]+$', '', name)      # remove trailing commas/spaces
    return name.strip()

# Apply to the DataFrame
wb_oz_df['product_name_second'] = wb_oz_df.apply(clean_product_name_full, axis=1)
wb_oz_df['product_name_second'].unique().tolist()
```

```python
# Inspect unique attributes' keys

from pprint import pprint
import json

unique_keys = set()
for row_idx, row in wb_oz_df.iterrows():
    items = row['OZAttributes (OZAttributes)']
    items = json.loads(items)
    for item in items:
        key = item['Item1']
        unique_keys.update([key])

print(f"total unique keys: {len(unique_keys)}\n")
pprint(unique_keys)
```

```python
# Extract necessary attributes' keys

import json

# Define the key(s) to extract
keys_to_extract = {'Пол'}

# Function to extract values from JSON string
def extract_attributes(json_str, keys):
    try:
        items = json.loads(json_str)
        extracted = {}
        for item in items:
            key = item.get('Item1')
            value = item.get('Item2')
            if key in keys:
                extracted[key] = value
        return extracted
    except (json.JSONDecodeError, TypeError):
        return {}

# Apply the extraction to each row and create new columns
extracted_df = wb_oz_df['OZAttributes (OZAttributes)'].apply(
    lambda x: extract_attributes(x, keys_to_extract)
).apply(pd.Series)

# Add extracted columns to the original DataFrame
for col in extracted_df.columns:
    wb_oz_df[col] = extracted_df[col]

wb_oz_df['Пол'].unique().tolist()
```

```python
# Normalize gender
gender_mapping = {
    'мальчики': 'для мальчика',
    'девочки': 'для девочки'
}

wb_oz_df['gender_norm_second'] = wb_oz_df['Пол'].str.lower()
wb_oz_df['gender_norm_second'] = wb_oz_df['gender_norm_second'].apply(
    lambda gender: gender_mapping.get(gender, gender)
)
wb_oz_df['gender_norm_second'].unique().tolist()
```

```python
# Concatenate name + gender

# Truncated forms of gender phrases to check presence
gender_substrings = {
    'для мальчика': ['для мальчик'],
    'для девочки': ['для девочк'],
    'мужской': ['мужск', 'для мужчин'],
    'женский': ['женск', 'для женщин', 'женские'],
}

def append_gender_if_missing_smart(name, gender):
    if pd.isna(gender):
        return name
    name_lower = name.lower()
    gender = gender.strip().lower()
    substrings = gender_substrings.get(gender, [])
    if not any(sub in name_lower for sub in substrings):
        return f"{name.strip()} {gender}"
    return name.strip()

wb_oz_df['name_second'] = wb_oz_df.apply(
    lambda row: append_gender_if_missing_smart(
        row['product_name_second'],
        row['gender_norm_second']
    ),
    axis=1
)

wb_oz_df['name_second'].unique().tolist()
```

# Shuffle

```python
DATA_PATH = 'data'
ds_path = Path(DATA_PATH) / "WB_OZ_100_conjugated.csv"

wb_oz_df_conjugated = pd.read_csv(ds_path, index_col=0)
wb_oz_df_conjugated.head()
```

```python
import pandas as pd
import numpy as np

# Assume wb_oz_df_conjugated is already loaded as a DataFrame

def shuffle_product_pairs(df, random_state=42, fraction=1):
    df = df.copy()
    
    # Determine the number of rows to shuffle
    num_shuffle = int(len(df) * fraction)
    indices_to_shuffle = np.random.RandomState(random_state).choice(df.index, size=num_shuffle, replace=False)
    
    # Shuffle only a fraction of the second product columns
    shuffled_df = df.loc[indices_to_shuffle].sample(frac=1, random_state=random_state).reset_index(drop=True)
    
    df.loc[indices_to_shuffle, 'name_second'] = shuffled_df['name_second']
    df.loc[indices_to_shuffle, 'description_second'] = shuffled_df['description_second']
    df.loc[indices_to_shuffle, 'image_name_second'] = shuffled_df['image_name_second']
    
    # Ensure no shuffled pair has identical image names
    mask = df.loc[indices_to_shuffle, 'image_name_first'] == df.loc[indices_to_shuffle, 'image_name_second']
    while mask.any():
        shuffled_df = df.loc[indices_to_shuffle].sample(frac=1, random_state=np.random.randint(0, 10000)).reset_index(drop=True)
        df.loc[indices_to_shuffle, 'name_second'] = shuffled_df['name_second']
        df.loc[indices_to_shuffle, 'description_second'] = shuffled_df['description_second']
        df.loc[indices_to_shuffle, 'image_name_second'] = shuffled_df['image_name_second']
        mask = df.loc[indices_to_shuffle, 'image_name_first'] == df.loc[indices_to_shuffle, 'image_name_second']
    
    # Update labels to 0 only for shuffled pairs
    df.loc[indices_to_shuffle, 'label'] = 0
    
    return df
```

```python
# Shuffle all

SEED = 42
FRACTION = 1

wb_oz_df_shuffled = shuffle_product_pairs(
    wb_oz_df_conjugated,
    random_state=SEED,
    fraction=FRACTION
)
wb_oz_df_shuffled.label.value_counts()
```

```python
filename = f"WB_OZ_100_conjugated_shuffled_seed={SEED}_fraction={FRACTION}.csv"
ds_path = Path(DATA_PATH) / filename
wb_oz_df_shuffled.to_csv(ds_path)
```
